```python
import json
import matplotlib.pyplot as plt

# Charger ton fichier JSON
with open("history.json", "r") as f:
    data = json.load(f)

# Accès à la section globale
global_data = data["global"]

# Rounds triés (les clés sont des strings → on les convertit)
rounds = sorted([int(r) for r in global_data.keys()])

# Accuracy correspondante
accuracies = [global_data[str(r)]["accuracy"] for r in rounds]

# Tracé
plt.plot(rounds, accuracies, marker='o')
plt.xlabel("Round")
plt.ylabel("Accuracy")
plt.title("Accuracy au fil des rounds")
plt.grid(True)
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Charger le fichier JSON
with open("history.json", "r") as f:
    data = json.load(f)

# Accès à la section post-fit
postfit_data = data["post-fit"]

# Extraire les rounds et clients
rounds = sorted([int(r) for r in postfit_data.keys()])

# Pour chaque round, calculer les statistiques des clients
mean_accuracies = []
std_accuracies = []
mean_f1 = []
std_f1 = []

for round_num in rounds:
    round_key = str(round_num)
    clients_data = postfit_data[round_key]
    
    # Extraire les accuracies et f1 de tous les clients de ce round
    accs = [clients_data[client]["accuracy"] for client in clients_data]
    f1s = [clients_data[client]["macro_f1"] for client in clients_data]
    
    mean_accuracies.append(np.mean(accs))
    std_accuracies.append(np.std(accs))
    mean_f1.append(np.mean(f1s))
    std_f1.append(np.std(f1s))

# Tracé avec 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Accuracy
ax1.plot(rounds, mean_accuracies, marker='o', linewidth=2, label='Mean')
ax1.fill_between(rounds, 
                 np.array(mean_accuracies) - np.array(std_accuracies),
                 np.array(mean_accuracies) + np.array(std_accuracies),
                 alpha=0.3, label='± 1 Std')
ax1.set_xlabel("Round")
ax1.set_ylabel("Accuracy")
ax1.set_title("Post-Fit Accuracy (Clients)")
ax1.grid(True, alpha=0.3)
ax1.legend()

# Macro F1
ax2.plot(rounds, mean_f1, marker='o', linewidth=2, label='Mean', color='green')
ax2.fill_between(rounds,
                 np.array(mean_f1) - np.array(std_f1),
                 np.array(mean_f1) + np.array(std_f1),
                 alpha=0.3, label='± 1 Std', color='green')
ax2.set_xlabel("Round")
ax2.set_ylabel("Macro F1")
ax2.set_title("Post-Fit Macro F1 (Clients)")
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Tracer quelques clients spécifiques
fig, ax = plt.subplots(figsize=(10, 6))

# Choisir 5 clients à afficher
selected_clients = ["66", "69", "95", "1", "52"]

for client_id in selected_clients:
    client_rounds = []
    client_accs = []
    
    for round_num in rounds:
        round_key = str(round_num)
        if client_id in postfit_data[round_key]:
            client_rounds.append(round_num)
            client_accs.append(postfit_data[round_key][client_id]["accuracy"])
    
    ax.plot(client_rounds, client_accs, marker='o', label=f'Client {client_id}')

ax.set_xlabel("Round")
ax.set_ylabel("Accuracy")
ax.set_title("Post-Fit Accuracy par Client")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```
