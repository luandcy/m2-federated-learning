```python
import yaml

# Charger la config existante
with open('config/exp.yaml', 'r') as f:
    config = yaml.safe_load(f)

# Configurer le logger pour sauvegarder en CSV
config['logger'] = {
    'name': 'Log',
    'save_dir': 'results/',  # Dossier de sauvegarde
    'save_metrics': True      # Activer la sauvegarde des métriques
}

# Sauvegarder la config modifiée
with open('config/exp.yaml', 'w') as f:
    yaml.dump(config, f)

print("Configuration du logger mise à jour !")
```

```python
from fluke import DDict
from fluke.run import run_experiment

# Charger les configurations
exp_config = DDict.load('config/exp.yaml')
fed_config = DDict.load('config/fedavg.yaml')

# Lancer l'expérience et récupérer le logger
logger = run_experiment(exp_config, fed_config)

# Sauvegarder l'historique du logger
logger.save('results/training_history.json')

print("Historique sauvegardé dans results/training_history.json")
```

```python
from ruamel.yaml import YAML

yaml = YAML()

with open("config/exp.yaml") as f:
    data = yaml.load(f)

# Activer la sauvegarde
data["logger"]["save"] = True

with open("config/exp.yaml", "w") as f:
    yaml.dump(data, f)

print("✔ Logger mis à jour : sauvegarde des métriques activée.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les métriques globales
global_metrics = pd.read_csv('results/global_metrics.csv')

# Afficher les premières lignes
print(global_metrics.head())

# Tracer l'accuracy globale au fil des rounds
plt.figure(figsize=(10, 6))
plt.plot(global_metrics['round'], global_metrics['accuracy'], marker='o')
plt.xlabel('Round')
plt.ylabel('Accuracy Globale')
plt.title('Évolution de l\'accuracy globale')
plt.grid(True)
plt.show()
```
